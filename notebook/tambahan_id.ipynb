```python
import pandas as pd

# Baca file CSV
df = pd.read_csv("data_final.csv")

# Tambahkan kolom 'id' dengan angka urut mulai dari 1
df.insert(0, 'id', range(1, len(df) + 1))

# Simpan ke file baru
df.to_csv("data_psikologi_with_id.csv", index=False)
```

```python
import pandas as pd
import uuid

# Baca file CSV
df = pd.read_csv("data_final.csv")

# Tambahkan kolom 'id' dengan UUID
df.insert(0, 'id', [str(uuid.uuid4()) for _ in range(len(df))])

# Simpan ke file baru
df.to_csv("data_psikologi_with_uuid.csv", index=False)
```
