```python
import pandas as pd
import os

# Folder tempat semua file CSV provinsi berada
folder_path = 'D:\DBS\Project Capstone\scrap data\data klinik bagian 1\hasil rename'

# List semua file dalam folder
all_files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]

# List untuk menyimpan semua DataFrame
df_list = []

# Looping untuk baca dan tambahkan kolom provinsi
for file in all_files:
    provinsi = file.replace('.csv', '').replace('_', ' ')  # Ambil nama provinsi dari nama file
    df = pd.read_csv(os.path.join(folder_path, file))      # Baca file CSV
    df['provinsi'] = provinsi                              # Tambahkan kolom provinsi
    df_list.append(df)                                     # Tambahkan ke list

# Gabungkan semua DataFrame
df_gabungan = pd.concat(df_list, ignore_index=True)
```
